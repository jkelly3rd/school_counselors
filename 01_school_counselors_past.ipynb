```python
import pandas as pd
```

```python
# import the statewide staffing data
staffing_2019 = pd.read_csv("data/ccd_state_staffing_091019.csv")

# show me the info and head of the data
staffing_2019.info()
staffing_2019.head()
```

```python
# import the statewide enrollment data
enrollment_2019 = pd.read_csv("data/ccd_state_membership_091019.csv")

# filter to just include education unit total in total_indicator column
enrollment_2019 = enrollment_2019[enrollment_2019["TOTAL_INDICATOR"] == "Education Unit Total"]

# filter to just include the school_year, statename, and student_count columns
enrollment_2019 = enrollment_2019[["SCHOOL_YEAR", "STATENAME", "STUDENT_COUNT"]]

# Eliminate records where STATENAME is in this list BUREAU OF INDIAN EDUCATION, AMERICAN SAMOA, GUAM, PUERTO RICO, U.S. VIRGIN ISLANDS
enrollment_2019 = enrollment_2019[~enrollment_2019["STATENAME"].isin(["BUREAU OF INDIAN EDUCATION", "NORTHERN MARIANAS", "AMERICAN SAMOA", "GUAM", "PUERTO RICO", "U.S. VIRGIN ISLANDS"])]

# show me the info and head of the data
enrollment_2019.info()
enrollment_2019.head(51)
```

```python
# Create a filter to create a new dataframe of only School Psychologists
school_psychologists_2019 = staffing_2019[staffing_2019["STAFF"] == "School Psychologists"]

# filter to just include the school_year, statename, and student_count columns
school_psychologists_2019 = school_psychologists_2019[["SCHOOL_YEAR", "STATENAME", "STAFF", "STAFF_COUNT"]]

# Merge with the enrollment data to get only STATENAME and STUDENT_COUNT columns
school_psychologists_2019 = pd.merge(school_psychologists_2019, enrollment_2019[["STATENAME", "STUDENT_COUNT"]], how="inner", on=["STATENAME"])

# Calculate a ratio of students to staff as a new column and round to one digit after decimal
school_psychologists_2019["STUDENT_TO_STAFF_RATIO"] = (school_psychologists_2019["STUDENT_COUNT"] / school_psychologists_2019["STAFF_COUNT"]).round(1)

# Show me the info and head of the data
school_psychologists_2019.info()
school_psychologists_2019.head(51)



```

```python
# Follow that same pattern to create a new dataframe of only Guidance Counselors
# This ensures that we're getting the derived TOTAL for separately-reported broken out counselors at all levels
school_counselors_2019 = staffing_2019[staffing_2019["STAFF"] == "Guidance Counselors"]

# filter to just include the school_year, statename, and student_count columns
school_counselors_2019 = school_counselors_2019[["SCHOOL_YEAR", "STATENAME", "STAFF", "STAFF_COUNT"]]

# Merge with the enrollment data to get only STATENAME and STUDENT_COUNT columns
school_counselors_2019 = pd.merge(school_counselors_2019, enrollment_2019[["STATENAME", "STUDENT_COUNT"]], how="inner", on=["STATENAME"])

# Calculate a ratio of students to staff as a new column and round to one digit after decimal
school_counselors_2019["STUDENT_TO_STAFF_RATIO"] = (school_counselors_2019["STUDENT_COUNT"] / school_counselors_2019["STAFF_COUNT"]).round(1)

# Show me the info and head of the data
school_counselors_2019.info()
school_counselors_2019.head(100)
```

```python
# Export csv files for both school psychologists and school counselors
school_psychologists_2019_2019.to_csv("output/school_psychologists_2019_2019.csv", index=False)
school_counselors_2019.to_csv("output/school_counselors_2019_2019.csv", index=False)
```
