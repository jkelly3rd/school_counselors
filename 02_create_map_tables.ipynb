```python
import pandas as pd
```

```python
# import school_counselors_2024.csv in output directory
school_counselors_2024 = pd.read_csv("output/school_counselors_2024.csv")
school_counselors_2020 = pd.read_csv("output/school_counselors_2020.csv")
school_psychologists_2024 = pd.read_csv("output/school_psychologists_2024.csv")
school_psychologists_2020 = pd.read_csv("output/school_psychologists_2020.csv")

# Merge school_counselors_2024 and school_counselors_2020 on the 'STATENAME' column
school_counselors_map_table = pd.merge(
    school_counselors_2024,
    school_counselors_2020[['STATENAME', 'STUDENT_TO_STAFF_RATIO']],
    on="STATENAME",
    suffixes=("_2024", "_2020"),
)

# Show me the info and head of the merged table
school_counselors_map_table.info()
school_counselors_map_table.head()
```

```python
# Merge school_counselors_2024 and school_counselors_2020 on the 'STATENAME' column
school_psychologists_map_table = pd.merge(
    school_psychologists_2024,
    school_psychologists_2020[['STATENAME', 'STUDENT_TO_STAFF_RATIO']],
    on="STATENAME",
    suffixes=("_2024", "_2020"),
)

# Show me the info and head of the merged table
school_psychologists_map_table.info()
school_psychologists_map_table.head(51)
```

```python
# Export both of the map chart files as csv to the output directory
school_counselors_map_table.to_csv(
    "output/school_counselors_map_table.csv", index=False
)
school_psychologists_map_table.to_csv(
    "output/school_psychologists_map_table.csv", index=False
)
```
