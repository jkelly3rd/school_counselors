```python
import pandas as pd
```

```python
# import the statewide staffing data
staffing_2024 = pd.read_csv("data/ccd_state_staffing_073124.csv")

# show me the info and head of the data
staffing_2024.info()
staffing_2024.head()
```

```python
# import the statewide enrollment data
enrollment_2024 = pd.read_csv("data/ccd_state_membership_073124.csv")

# filter to just include education unit total in total_indicator column
enrollment_2024 = enrollment_2024[enrollment_2024["TOTAL_INDICATOR"] == "Education Unit Total"]

# filter to just include the school_year, statename, and student_count columns
enrollment_2024 = enrollment_2024[["SCHOOL_YEAR", "STATENAME", "STUDENT_COUNT"]]

# Eliminate records where STATENAME is in this list BUREAU OF INDIAN EDUCATION, AMERICAN SAMOA, GUAM, PUERTO RICO, U.S. VIRGIN ISLANDS
enrollment_2024 = enrollment_2024[~enrollment_2024["STATENAME"].isin(["BUREAU OF INDIAN EDUCATION", "NORTHERN MARIANAS", "AMERICAN SAMOA", "GUAM", "PUERTO RICO", "U.S. VIRGIN ISLANDS"])]

# show me the info and head of the data
enrollment_2024.info()
enrollment_2024.head(51)
```

```python
# Create a filter to create a new dataframe of only School Psychologists
school_psychologists = staffing_2024[staffing_2024["STAFF"] == "School Psychologists"]

# filter to just include the school_year, statename, and student_count columns
school_psychologists = school_psychologists[["SCHOOL_YEAR", "STATENAME", "STAFF", "STAFF_COUNT"]]

# Merge with the enrollment data to get only STATENAME and STUDENT_COUNT columns
school_psychologists = pd.merge(school_psychologists, enrollment_2024[["STATENAME", "STUDENT_COUNT"]], how="inner", on=["STATENAME"])

# Show me the info and head of the data
school_psychologists.info()
school_psychologists.head(51)



```

```python
# Follow that same pattern to create a new dataframe of only Guidance Counselors
# This ensures that we're getting the derived TOTAL for separately-reported broken out counselors at all levels
school_counselors = staffing_2024[staffing_2024["STAFF"] == "Guidance Counselors"]

# filter to just include the school_year, statename, and student_count columns
school_counselors = school_counselors[["SCHOOL_YEAR", "STATENAME", "STAFF", "STAFF_COUNT"]]

# Merge with the enrollment data to get only STATENAME and STUDENT_COUNT columns
school_counselors = pd.merge(school_counselors, enrollment_2024[["STATENAME", "STUDENT_COUNT"]], how="inner", on=["STATENAME"])

# Show me the info and head of the data
school_counselors.info()
school_counselors.head(100)
```
